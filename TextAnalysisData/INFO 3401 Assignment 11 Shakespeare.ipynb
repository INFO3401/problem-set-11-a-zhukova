## Problem 5

```python
df_raw = pd.read_csv('rawCounts.csv')
df_raw
```

```python
print(df_raw.columns)
```

```python
df_groupby = df_raw.groupby(['Author', 'Word']).sum()
df_groupby
```

```python
sort = df_groupby.reset_index().sort_values(['Author', 'Count'], ascending = False).set_index(['Author', 'Word'])
sort
```

```python
top = sort.groupby('Author').head(2500).reset_index()
top
```

## Problem 6

This visualiztaion shows that Marlowe's word counts almost exlusively fall below 4000 while Shakespeare has wordcounts that fall above 4000. This tells us that Shakepeare was likely more verbose that Marlowe in writing his plays and that his plays were likely longer.

```python
alt.Chart(top).mark_point().encode(
    x='Author',
    y='Count',
).interactive()
```

The below graphic shows that the words that both playwrites used the most in the plays were pronouns and conjunctions almost equally, which makes sense considering they are a constructive part of English language. This doesn't seem to imply any significant relationship about their collaboration.

```python
top_500 = sort.groupby('Author').head(500).reset_index()
alt.Chart(top_500).transform_window(
    rank='rank(Count)',
    sort=[alt.SortField('Count', order='descending')]
).transform_filter(
    alt.datum.rank < 100
).mark_bar().encode(
    y=alt.Y('Word:N',
        sort=alt.EncodingSortField(field='Count', op='sum', order='descending')
    ),
    x='Count:Q',
    color = 'Author'
)
```

This chart seems to show that Marlowe's Hero and Leander play has the most positive sentiment out of all the plays as opposed to negative and Shakespeare's Endward II has the most equal ratio of positive to negative sentiment.
